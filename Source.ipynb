```python
import IPython
from scipy.io import wavfile
import scipy.signal
import numpy as np
import matplotlib.pyplot as plt
import librosa
import noisereduce as nr
import tensorflow
import soundfile
import glob
import pandas as pd
%matplotlib inline
```

```python
def NReAL_engine(data,rate):
    #rate, data = wavfile.read(filepath)
    #fig, ax = plt.subplots(figsize=(20,4))
    #ax.plot(data)
    #print(data)
    #IPython.display.Audio(data=data, rate=rate)
    out = lgm.logmmse(data = data,sampling_rate = rate,output_file="out.wav",initial_noise=6,window_size=0,noise_threshold=0.15)
    #print(out)
    #fig, ax = plt.subplots(figsize=(20,4))
    #ax.plot(out)
    #IPython.display.Audio(data=out, rate=rate)
    #applying noise reduction
    out = out/32768
    noisy_part = out[0:400000]
    # perform noise reduction
    reduced_noise = nr.reduce_noise(audio_clip=out,noise_clip= noisy_part,verbose=False,prop_decrease = 1.0)
    #IPython.display.Audio(data=reduced_noise, rate=rate)
    #re-enhancement / final output
    out = lgm.logmmse(data = reduced_noise,sampling_rate = rate,initial_noise=6,window_size=0,noise_threshold=0.15)
    #print(out)
    #IPython.display.Audio(data=out, rate=rate)
    #Plot of the output file
    #fig, ax = plt.subplots(figsize=(20,4))
    #ax.plot(out*32768)
    return out
```

```python
#MALE F0 CALCULATIONS
m_mean_F0_list = []
mfile_list = []
mcount = 0
for wave_file in glob.glob("Data/Male/*.wav"):
    rate, data = wavfile.read(wave_file)
    #print(wave_file ,"\t",sig.get_F_0(data,rate))
    mean = sig.get_F_0(data,rate)
    mcount += 1
    m_mean_F0_list.append(mean)
    mfile_list.append(wave_file)     
```

```python
gender = ["M"]*mcount 
male_df = pd.DataFrame(np.column_stack([mfile_list, m_mean_F0_list,gender]), 
                               columns=['voiceID', 'meanF0Hz','Gender']) 
#male_df
```

```python
f0_sum = np.sum(m_mean_F0_list)
avg_male_f0 =  f0_sum / mcount
print(avg_male_f0)
print(male_df)
```

```python
#FEMALE F0 CALCULATIONS
```

```python
f_mean_F0_list = []
ffile_list = []
fcount = 0
for wave_file in glob.glob("Data/Female/*.wav"):
    rate, data = wavfile.read(wave_file)
    #print(wave_file ,"\t",sig.get_F_0(data,rate))
    mean = sig.get_F_0(data,rate)
    fcount += 1
    f_mean_F0_list.append(mean)
    ffile_list.append(wave_file)
```

```python
gender = ["F"]*fcount 
female_df = pd.DataFrame(np.column_stack([ffile_list, f_mean_F0_list,gender]), 
                               columns=['voiceID', 'meanF0Hz','Gender']) 
```

```python
f0_sum = np.sum(f_mean_F0_list)
avg_female_f0 =  f0_sum / fcount
print(avg_female_f0)
print(female_df)
```

```python
#Testing MALE prediction
```

```python
mmean_F0_list = []
mfile_list = []
mprediction = []
mcount = 0
mpcount = 0
for wave_file in glob.glob("Data/Male/*.wav"):
    rate, data = wavfile.read(wave_file)
    #print(wave_file ,"\t",sig.get_F_0(data,rate))
    mean = sig.get_F_0(data,rate)
    mcount += 1
    mmean_F0_list.append(mean)
    mfile_list.append(wave_file)
    if abs(mean[0] - avg_female_f0) < abs(mean[0] - avg_male_f0):
        mprediction.append("F")
    else:
        mprediction.append("M")
        mpcount += 1
    
```

```python
gender = ["M"]*mcount 
mpdf = pd.DataFrame(np.column_stack([mfile_list, mmean_F0_list,gender,mprediction]), 
                    columns=['voiceID', 'meanF0Hz','ActualGender','Prediction'])
print(mpdf)
```

```python
m_acc = (mpcount / mcount) * 100
print("Male Accuracy : ",m_acc," %")
```

```python
#Testing FEMALE prediction
```

```python
fmean_F0_list = []
ffile_list = []
fprediction = []
fcount = 0
fpcount = 0
for wave_file in glob.glob("Data/Female/*.wav"):
    rate, data = wavfile.read(wave_file)
    mean = sig.get_F_0(data,rate)
    #print(wave_file ,"\t",mean)
    fcount += 1
    if mean[0] < 75 : 
        pass
    fmean_F0_list.append(mean)
    ffile_list.append(wave_file)
    if abs(mean[0] - avg_female_f0) < abs(mean[0] - avg_male_f0):
        fprediction.append("F")
        fpcount += 1
    else:
        fprediction.append("M")
        
    
```

```python
gender = ["F"]*fcount 
fpdf = pd.DataFrame(np.column_stack([ffile_list, fmean_F0_list,gender,fprediction]), 
                               columns=['voiceID', 'meanF0Hz','ActualGender','Prediction']) 
print(fpdf)
```

```python
f_acc = (fpcount / fcount) * 100
print("Female Accuracy",f_acc, "%")
```

```python
#COMBINING THE TEST RESULTS
```

```python
df_row_reindex = pd.concat([mpdf, fpdf], ignore_index=True)
df_row_reindex = df_row_reindex.sample(frac = 1).reset_index(drop=True)
df_row_reindex.to_csv("results.csv",index = False)
print(df_row_reindex)
```

```python
#CHECKING ACCURACY ON COMBINED RESLUTS
match = 0 
total = 1838
for i in range (len (df_row_reindex)):
    if df_row_reindex.loc[i,"ActualGender"] == 'F' and df_row_reindex.loc[i,"Prediction"] == 'F':
        match+=1
    elif df_row_reindex.loc[i,"ActualGender"] == 'M' and df_row_reindex.loc[i,"Prediction"] == 'M' :
        match+=1
```

```python
accuracy = 100 * match/total
accuracy
```
